```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# ==========================================
# 1. Create the Excel Data Programmatically (8 independent rows)
# ==========================================
data = {
    'Class': ['Normal', 'Normal', 'Normal', 'Normal', 'Spam', 'Spam', 'Spam', 'Spam'],
    'Word': ['Dear', 'Friend', 'Lunch', 'Money', 'Dear', 'Friend', 'Lunch', 'Money'],
    'Word Count': [6, 3, 2, 1, 3, 3, 1, 1]
}

df = pd.DataFrame(data)
print("--- Loaded Dataset (Simulated Excel) ---")
print(df)
print("-" * 40)

# ==========================================
# 2. Advanced Preprocessing WITHOUT Groupby
# ==========================================
# Replicate each word in its own separate row to preserve dataset size and avoid 0.00 accuracy
df['Weighted_Word'] = df.apply(lambda row: (str(row['Word']) + ' ') * int(row['Word Count']), axis=1)

# Features (X) and Labels (y) from individual rows (Keeping 8 samples)
X = df['Weighted_Word']
y = df['Class']

# ==========================================
# 3. Vectorization & Feature Extraction
# ==========================================
vectorizer = CountVectorizer()
X_vectorized = vectorizer.fit_transform(X)

# ==========================================
# 4. Train-Test Split & Model Training
# ==========================================
# Using a 25% split (2 rows for testing, 6 rows for training)
# This ensures both classes exist in both training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.25, random_state=42)

model = MultinomialNB()
model.fit(x_train, y_train)

# ==========================================
# 5. Model Evaluation (Correct and Realistic)
# ==========================================
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"\nAccuracy: {accuracy:.2f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))
print("-" * 40)

# ==========================================
# 6. Predict on New Text
# ==========================================
new_text = ["Dear Friend Lunch"]
new_text_transformed = vectorizer.transform(new_text)
prediction = model.predict(new_text_transformed)
print(f"The predicted class for {new_text} is: {prediction[0]}")
```
